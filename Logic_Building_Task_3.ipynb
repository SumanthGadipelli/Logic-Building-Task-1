1. Employee Performance Bonus Eligibility

```python
emp={"Ravi":92, "Anita":88, "Kiran":92, "Suresh":85} # Given Dicitionary of employees & their score
high_score=0
for score in emp.values(): # To find highest score
    if score>high_score:
        high_score=score
top_performers=[] # To find employees with highest score
for name, score in emp.items():
    if score==high_score:
        top_performers.append(name)
names_string=",".join(top_performers)
print(f"Top performers Eligible for Bonus: {names_string} (Score: {high_score})")
```

2. Search Query Keyword Analysis

```python
search_query="Buy mobile phone buy phone online" # Given Search Query
cleaned=search_query.lower() # To Convert to Lower case
punctuations = ".,!?;:" # To Remove Punctuations
for char in punctuations:
  cleaned=cleaned.replace(char,"")
words=cleaned.split()
word_count={}
for word in words: # To Count Frequency of each word
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word]=1
repeat_key={} # To Filter Keywords searched more than once
for word, count in word_count.items():
  if count>1:
    repeat_key[word]=count
print(repeat_key)
```

3. Sensor Data validation

```python
sensor_reading=[3, 4, 7, 8, 10, 12, 5] # Given Sensor Readings
valid=[] # List to store valid readings
for hours, reading in enumerate(sensor_reading): # To iterate through list
  if reading % 2 == 0:
    valid.append((hours, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

4. Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"] # Given list of emails

domain_counts={}
for email in emails: # To separate name and domain
    parts=email.split("@")
    domain=parts[1]
    if domain in domain_counts: # To update Domain count
        domain_counts[domain]+=1
    else:
        domain_counts[domain]=1
total_users=len(emails)
for domain, count in domain_counts.items():
    percentage=(count / total_users) * 100 # Calculate percentage
    print(f"{domain}: {int(percentage)}%")
```

5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2400, 1400, 3000] # Given Sales list
total=0
for sale in sales: # Toatal Sales
    total+=sale
avg=total/len(sales) # Average Sales
threshold=avg+(0.3*avg) # To Find Days where Sales are more than 30% above average
day=1
for sale in sales:
    if sale>threshold:
        print("Day", day, ":", sale)
    day=day+1
```

6. Duplicate User ID Detection

```python
user_ids=["user1", "user2", "user1", "user3", "user1", "user3"] # Given User ID's
id_counts={}
for user in user_ids: # To Count the occurences of each userID
    if user in id_counts:
        id_counts[user]+=1
    else:
        id_counts[user]=1
for user, count in id_counts.items(): # To Display only ID's that appear more than once
    if count > 1:
        print(f"{user} → {count} times")
```
